```python
def mcdonals(dictrandom):
   names=[]
   lats=[]
   long=[]
   for place in dictrandom["results"]:
       names.append(place["name"])
       lats.append(place["geometry"]["location"]["lat"])
       long.append(place["geometry"]["location"]["lng"])
   dictmcdonals = {"name":names, "latitude":lats, "longitude":long}
   return pd.DataFrame(dictmcdonals)
mcdonals = mcdonals(mc)
mcdonals.head()
```

```python
import folium
import pandas as pd
from folium.plugins import HeatMap
import requests
from folium.plugins import MeasureControl
```

```python
map_mcdonals = folium.Map(location=[37.781265, -122.393229], width=750, height=500, zoom_start=15)
for index, row in mcdonals.iterrows():
    
    folium.CircleMarker([row['latitude'], row['longitude']],
                       radius=9,
                       popup="latitude {}, longitude {}".format(row['latitude'], row['longitude']),
                       fill_color="#F35C40",
                      ).add_to(map_mcdonals)
    folium.Marker(location=[37.781265, -122.393229],
                   radius=100
                  ).add_to(map_mcdonals)
map_mcdonals.add_child(MeasureControl())
map_mcdonals.save('./map_mcdonals.html') 
map_mcdonals
```
